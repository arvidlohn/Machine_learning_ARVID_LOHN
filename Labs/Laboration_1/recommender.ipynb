```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df_movies_new_new = df_movies_new_new.reset_index(drop=True)
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df_movies_new_new["features"])
```

```python
tfidf_matrix.shape
```

```python
indices = pd.Series(
    df_movies_new_new.index,
    index = df_movies_new_new["title"]
).drop_duplicates()
```

```python
indices["Elevator Movie (2004)"]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(movie_title, n=5):
    idx = indices[movie_title]

    similarities = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    similar_indices = similarities.argsort()[::-1]
    similar_indices = similar_indices[1:n+1]

    return df_movies_new_new.iloc[similar_indices][["title", "genres"]]
```

```python
recommend("Toy Story (1995)")
```

```python
recommend("Jumanji (1995)")
```
