# Movie Recommendation System

### Introduction
Denna uppgift syftar på att bygga ett rekomentationssystem för filmer med datasetet MovieLens-dataset.
Systemet ska kunna ta emot en film title som input och sedan då rekomendera fem likande filmer.

Jag valde att bygga ett rekomendationsystem som är innehållsbaserat. Det innebär att rekomendationerna baseras på information om filmerna själva. I mitt fall tog jag genre och tags. Tags är ju användarinput. 

Metoden jag valde att använda är TF-IDF tillsammans med cosine similarity.

### Dataset
Jag fick en massa dataset. 
 - movies.csv
 - ratings.csv
 - tags.csv
 - links.csv

I min modell som jag byggde använde jag mig av movies.csv samt tags.csv.

movies.csv innehåller infromation om filmernas genre och title. 
tags.csv innehåller taggar som anvädnare/filmkollare har kopplat till olika filmer. 

Min slutasts när jag gick igenom datan var att taggarna som användare satt på filmerna känndes väldigt rimliga därav valde jag att ha med taggarna i rekomendationsystemt. 

### Data Preparation
Genrens i datasetet movies.csv var separerat med | tecknet mellan varje genre som filmen hade.

Detta tog jag bort så | ersattes med ett mellanrum istället. Detta för att systemet inte skulle hänga upp sig på tecken utan enbart orden.
Detta gjordes även så att genres kunde behandlas som vanlig text.

Sedan valde jag att gruppera taggarna som låg i tags.csv till movieID, så att varje film fick en samlad textsräng med alla sina taggar.
Saknades det taggar på en film fick det en tom sträng. 
Jag valde att ta bort alla filmer som saknade både genre och taggar då jag inte kunde använda dessa i mitt rekomendationssystem. Jag valde även att skapa en till kolumn features där genre och tags kolumnerna kombinerades alltså blev till 1 kolumn. 

### Method
Jag valde att använda en innehållsbaserad metod. 

Systemet jämför filmer med varandra baserat på deras innehåll, alltså genre och taggar i detta fallet. 

För att omvandla text till siffror användes TfidfVectorizer. Denna rekomendation fick jag från CHATGPT att använda denna metod.

TF-IDF fokuserar ju mer på eller ger högre betydelse till ord som inte förekommer mycket i datasetet. Det vill säga om ett ord förekommer mycket i genre eller i taggarna så tycker TF-IDF inte är det är så intressant utan den "letar" mer efter ord som inte förekommer så ofta. 

Efter att filmerna hade omvandlats till numersika vektorer använde jag cosine similarity för att mäta hur lika filmerna är varnadra.
De filmerna som fick högst likhetspoäng rekomenderade systemet. 


### Implementation
Systemet är uppdelat i flera funktioner.

1. load_data()
Läser in datasetet från angiven mapp.

2. data_prep()
Förbereder datan genom att rensa taggar, gruppera taggar per film, rensa genre och skapa kolumnen features.

3. build_model()
Skapar TF-IDF-matrisen och ett index som kopplar filmtitlar till rätt rad i df.

4. recommend()
Tar emot en filmtitle och retunerar de mest liknande filmerna. 

5. main()
Gör att systemet kan köras från terminal.

Exempel:
python recommender.py "Toy Story (1995)"

Om man skulle vilja kan man även få välja hur måånga rekomendationer man vill ha defult är inställt på 5 men vill man ha mer går det bra också.
Då skriver man på följande sätt:
python recommender.py "Toy Story (1995)" -n 10

### Limitations
Rekomendationsystemet har några begränsningar.

Först och främst används inte användarbetygen från ratings.csv. Det betyder att modellen inte lär sig av vad olika använader faktiskt tycker om filmerna.

För det andra beror resultatet mycket på kvaliteten på taggarna. Om en film har få eller dåliga taggar kan rekomendationsystsmet rekomendera sämre.

För det tredje rekomenderar modellen oftast filmer som är väldigt lika inputfilmen. Detta kan göra dem rekomenderade filmerna mindre varierande.

Modellen förstår inte heller text på djupet. Utan jämför oftast ordmönster med hjälp av TF-IDF och cosine similarity.
